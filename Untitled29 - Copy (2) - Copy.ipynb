```python
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

import matplotlib.pyplot as plt
```

```python
cancer=load_breast_cancer()
X_train, X_test, y_train, y_test=train_test_split(cancer.data, cancer.target, random_state=0)

mlp=MLPClassifier(random_state=42)
mlp.fit(X_train, y_train)
```

```python
print('Accuracy on trainig set:{:.3f}'.format(mlp.score(X_train, y_train)))
```

```python
print('The maximum per feature\n{}'.format(cancer.data.max(axis=0)))
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
X_train_scaled=scaler.fit(X_train).transform(X_train)
X_test_scaled=scaler.fit(X_train).transform(X_test)

mlp=MLPClassifier(max_iter=1000, random_state=42)
mlp.fit(X_train_scaled, y_train)

print('Accuracy on the training set:{:.3f}'.format(mlp.score(X_train_scaled, y_train)))
print('Accuracy on the test set:{:.3f}'.format(mlp.score(X_test_scaled, y_test)))
```

```python
mlp=MLPClassifier(max_iter=1000, alpha=1, random_state=42)
mlp.fit(X_train_scaled, y_train)
print('Accuraacy on the training set:{:.3f}'.format(mlp.score(X_train_scaled, y_train)))
print('Accuraacy on the test set:{:.3f}'.format(mlp.score(X_test_scaled, y_test)))
```

```python
plt.figure(figsize=(20,5))
plt.imshow(mlp.coefs_[0], interpolation='None', cmap='GnBu')
plt.yticks(range(30), cancer.feature_names)
plt.xlabel('Columns in wt matrix')
plt.ylabel('input feature')
plt.colorbar()


```
