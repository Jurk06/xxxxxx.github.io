```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-6,6,50)

cos=np.cos(x)
sin=np.sin(x)
plt.plot(x, cos, linestyle='--', linewidth=3.5, color='g')
plt.plot(x, sin,linestyle='-', linewidth=3, color='b')
plt.text(-4,-0.3, 'This is text', rotation=10, fontsize=14)
plt.xlabel("X_axis")
plt.ylabel("Y_axis")
plt.title("Sine+cose")


plt.show()
```

```python
x=np.linspace(-10,10,50)
plt.subplot(211)
cos=np.cos(x)
sin=np.sin(x)
plt.plot(x, cos, linestyle='--', linewidth=3.5, color='g')
plt.subplot(212)
plt.plot(x, sin,linestyle='-', linewidth=3, color='b')
plt.text(-4,-0.3, 'This is text', rotation=10, fontsize=14)
plt.xlabel("X_axis")
plt.ylabel("Y_axis")
plt.title("Sine+cose")
plt.xlim(-15,15)
plt.ylim(-2,2)

plt.show()

```
